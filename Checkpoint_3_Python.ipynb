# **Paso 24:** Distribuciones Muestrales y T√©cnicas de Evaluaci√≥n de Modelos

- Sesgos y  muestreo aleatorio

**Nota:** Cuando tenemos una muestra que no es representativa (por ejemplo con 90% de hombres y 10% de mujeres) tenemos un conjunto de datos sesgado. Un sesgo es una tendencia de nuestros datos. En este caso es una tendencia hacia favorecer la existencia de hombres en nuestro conjunto. Normalmente los sesgos van en contra de la racionalidad y en el caso de la estad√≠stica, los sesgos van en contra de nuestra capacidad de obtener informaci√≥n acerca de una poblaci√≥n a partir de una muestra.
No siempre podemos deshacernos de un sesgo en nuestros datos, pero al menos es muy importante saber que el sesgo est√° presente. Hay algunas maneras en las que podemos resguardarnos de los sesgos. Una de ellas es el muestreo aleatorio (aleatorio).
Un demostrado aleatoriamente implica tomar elementos de nuestra poblaci√≥n de manera aleatoria con el fin de generar una muestra. Cada elemento debe de tener la misma probabilidad de ser seleccionado, con el fin de evitar en lo posible la creaci√≥n de muestras sesgadas.
‚Ä¢	Podemos realizar el probado sin reposici√≥n de elementos, es decir: cada vez que extraemos un elemento de la poblaci√≥n no lo ponemos de regreso.
‚Ä¢	Pero tambi√©n podemos realizar el muestreo con reposici√≥n, es decir: despu√©s de extraer un elemento y agregarlo a la muestra, el elemento se pone de regreso en la poblaci√≥n o muestra original. En este caso, podemos tener elementos repetidos en la muestra resultante.

- Aplicaci√≥n del algoritmo Bootstrap y obtenenci√≥n de errores est√°ndares e intervalos de confianza.
  
**Nota:**  Normalmente es muy dif√≠cil (o imposible) regresar al origen de nuestra muestra para extraer m√°s muestras y poder compararlas con otras. Adem√°s, normalmente queremos aprovechar la mayor√≠a de los datos que tenemos para obtener una medida estad√≠stica precisa. Es por eso que se usan algunas t√©cnicas que nos permiten "fingir" este re-muestreo utilizando solamente los datos que tenemos a la mano. Una de esas t√©cnicas es la llamada bootstrap.
La t√©cnica de bootstrapping se utiliza para obtener distintas muestras a partir de la muestra que tenemos en nuestras manos y obtener un valor que nos indica qu√© tanta incertidumbre hay en la medida estad√≠stica que hemos realizado. Podr√≠as pensarlo como: "Si pudiera regresar al origen de mis datos y tomar m√°s muestras de mi poblaci√≥n, ¬øqu√© diferencia esperar√≠a obtener en el c√°lculo de esta medida estad√≠stica entre las diferentes muestras?". Usando el boostrap te puede dar una idea de qu√© tanto puedes confiar en tu medida estad√≠stica.
Los pasos para realizar un boostrap son los siguientes:
‚Ä¢	Toma un elemento de tu conjunto de datos de manera aleatoria con reposici√≥n.
‚Ä¢	Repite el paso 1 n veces (entre m√°s cerca est√© na la longitud total de tu muestra, mejor).
‚Ä¢	Toma la medida estad√≠stica que te interesa de tus valores remuestreados.
‚Ä¢	Repite los pasos 1 a 3 R veces (entre mayor sea R, mejor).
‚Ä¢	Utilice las medidas obtenidas para:
a) Generar un histograma o boxplot
b) Calcular el error est√°ndar
c) Calcular un intervalo de confianza
si la distribuci√≥n obtenida diverge mucho de la distribuci√≥n normal podemos hacer 3 suposiciones:
‚Ä¢	O nuestra muestra original est√° muy sesgada y resulta en una distribuci√≥n asim√©trica.
‚Ä¢	Hemos realizado un n√∫mero muy peque√±o de pruebas y por lo tanto no tenemos una estimaci√≥n precisa.
‚Ä¢	O las muestras que estamos tomando son muy peque√±as y facilitan la aparici√≥n de sesgos en nuestra distribuci√≥n resultante.
